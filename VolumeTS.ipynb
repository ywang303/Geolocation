```python
# top 20 by total volume
top20list = l_1[0:20]
```

```python
top20df = [df.loc[df['RC_STATION'] == top20list[i], l] for i in range(20)]

days = [top20df[i].iloc[0] for i in range(20)]
dayslist = [days[i].values.tolist() for i in range(20)]
```

```python
for i in range(20):
    l = dayslist[i]
    volumeList = []
    for j in range(len(l)):
        if(l[j].isspace()):
            volumeList.append(0)
        else:
            volumeList.append(int(l[j]))
    plt.plot(volumeList)
    plt.show()
```
